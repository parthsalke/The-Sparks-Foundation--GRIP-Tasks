```python
#Renaming the columns
data.rename(columns={'iyear':'Year','imonth':'Month','iday':"Day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
```

```python
#extracting import attributes from dataset
data = data[['Year','Month','Day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive']]
```

```python
#checking null values
data.isnull().sum()
```

```python
#filling 0 in Null Values
data['Wound'] = data['Wound'].fillna(0)
data['kill'] = data['kill'].fillna(0)
```

```python
data.isnull().sum()
```

```python
data['Casualities'] = data['kill'] + data['Wound']
```

```python
data.info()
```

```python
#Most Affected countries
attack = data.Country.value_counts()[:15]
attack
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(data['Country'].value_counts()[:15].index,data['Country'].value_counts()[:15].values)
plt.title('Top 15 Affected Countries', fontsize=23)
plt.xlabel('Countries', fontsize=18)
plt.ylabel('Count', fontsize=18)
plt.xticks(rotation = 50)
plt.show()
```

```python
#Terrorism groups and attacks
df=data[['Group','Country','kill']]
df=df.groupby(['Group','Country'],axis=0).sum().sort_values('kill',ascending=False).drop('Unknown').reset_index().head(10)
df
```

```python
# plot number of people died in the attack
df = data[['Year','kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,10))
df.plot(kind='bar',alpha=0.8,ax=ax4)
plt.xticks(rotation = 50)
plt.title("People Died By Attack",fontsize=25)
plt.ylabel("Number of people killed",fontsize=20)
plt.xlabel('Year',fontsize=20)
top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

```python
# Area plot for finding attackes by region in each year
pd.crosstab(data.Year, data.Region).plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Terrorist Activities By Region In Each Year',fontsize=25)
plt.ylabel('Number of Attacks',fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```

```python
#plot for finding number of attackes in each year
year = data['Year'].unique()
years_count = data['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = year, y = years_count)
plt.xticks(rotation = 50)
plt.xlabel('Attacking Year',fontsize=20)
plt.ylabel('Number of Attacks Each Year',fontsize=20)
plt.title('Attacks In Years',fontsize=25)
plt.show()
```

```python
# Plot number of people killed in different attackes
data[['Attacktype','kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,10),color='purple')
plt.xticks(rotation=50)
plt.title("Number of people killed in different attackes ",fontsize=25)
plt.ylabel('Number of people',fontsize=20)
plt.xlabel('Attack type',fontsize=20)
plt.show()
```

```python
#Number of attackes by different ways
data['Attacktype'].value_counts().plot(kind='bar',figsize=(10,10),color='orange')
## plt.xticks(rotation = 50)
plt.xlabel("TYPE OF ATTACK",fontsize=20)
plt.ylabel("NUMBER OF ATTACK",fontsize=20)
plt.title("Number of attackes by different ways",fontsize=25)
plt.show()
```

```python
# Top 15 Terrorist Group in these attackes 
data[['Group','kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('kill',ascending=False).head(15).plot(kind='bar',figsize=(10,10))
plt.title("Top 15 terrorist group in these attackes",fontsize=25)
plt.xlabel("Terrorist group name",fontsize=20)
plt.ylabel("Number of people killed",fontsize=20)
plt.show()
```

```python
#Top 15 affected City from the attack
data['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(15).plot(kind='bar',figsize=(10,10), color='red')
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=20)
plt.ylabel("Number of attack",fontsize=20)
plt.title("Top 15 Most Affected Cities",fontsize=25)
plt.show()
```

```python
#Top affected Regions from the attack
data['Region'].value_counts().to_frame().sort_values('Region',axis=0,ascending=False).head(20).plot(kind='bar',figsize=(10,10))
plt.xticks(rotation = 50)
plt.xlabel("Regions",fontsize=20)
plt.ylabel("Number of attacks",fontsize=20)
plt.title("Top Affected Regions",fontsize=25)
plt.show()
```
