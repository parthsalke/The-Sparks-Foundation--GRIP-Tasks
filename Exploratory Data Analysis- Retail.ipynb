```python
sns.countplot(x=data['Region'])
```

```python
plt.figure(figsize=(10,7))
sns.countplot(x='State',data=data,order=data['State'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of orders in each State')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.countplot(x='Sub-Category',data=data,order=data['Sub-Category'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of items in Sub-Category')
plt.show()
```

```python
plt.figure(figsize=(10,8))
data['Region'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

```python
data['Quantity'].value_counts()
```

```python
data.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['red','orange'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
data.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['blue','red'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
ps = data.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['Orange','lightblue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

```python
ps = data.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['blue','orange'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

```python
sns.lineplot(x='Quantity',y='Discount',label='Quantity',data=data)
plt.legend()
plt.show()
```
